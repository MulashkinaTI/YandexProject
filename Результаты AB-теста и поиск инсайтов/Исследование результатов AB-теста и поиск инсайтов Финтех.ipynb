#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.2.** Анализ воронок событий.
- Постройте классическую и пошаговую воронки;
- Охарактеризуйте пользовательский путь в приложении и определите узкие места воронки. 


````python
funnel_steps = ['install / open_web', 'introduction', 'registration', 'main_page',
 'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']
funnel_df_steps = df_hist[df_hist['event_name'].isin(funnel_steps)]
funnel_counts = funnel_df_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)
funnel_df = pd.DataFrame({
    'step': funnel_steps,
    'users': funnel_counts.values})
funnel_df['conversion_from_first_%'] = round((funnel_df['users'] / funnel_df['users'].iloc[0]) * 100, 1)
funnel_df['conversion_from_previous_%'] = round((funnel_df['users'] / funnel_df['users'].shift(1)) * 100, 1) 


plt.figure(figsize = (20, 20))

plt.subplot(3, 1, 1)
bars = plt.bar(funnel_df['step'][1:], funnel_df['conversion_from_first_%'][1:])
plt.title('Воронка: конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.ylabel('Конверсия от первого шага в %')

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height}',
        ha='center',
        va='bottom',
        fontsize=10
        )

plt.subplot(3, 1, 2)

bars = plt.bar(funnel_df['step'][1:], funnel_df['conversion_from_previous_%'][1:])

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height}',
        ha='center',
        va='bottom',
        fontsize=10
        )

plt.title('Step-by-step воронка: конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.ylabel('Конверсия от предыдущего шага в %')

plt.show()
````

````python
mobile_df = df_hist[df_hist['platform'] == 'mobile']
web_df = df_hist[df_hist['platform'] == 'web']

mobile_df_steps = mobile_df[mobile_df['event_name'].isin(funnel_steps)]
mobile_counts = mobile_df_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)
mobile_funnel = round((mobile_counts / mobile_counts.shift(1) * 100), 1)

web_df_steps = web_df[web_df['event_name'].isin(funnel_steps)]
web_counts = web_df_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)
web_funnel = round((web_counts / web_counts.shift(1) * 100), 1)


funnel_device = pd.DataFrame({
    'step': funnel_steps[1:],
    'users_mobile': mobile_funnel[1:].values,
    'users_web': web_funnel[1:].values})
 
plt.figure(figsize = (20, 20))

plt.subplot(3, 1, 1)
bars = plt.bar(funnel_device['step'], funnel_device['users_mobile'])
plt.title('Step-by-step воронка: конверсия от предыдущего шага (%) для мобильной версии')
plt.xlabel('Этапы воронки')
plt.ylabel('Конверсия от предыдущего шага в %')

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height}',
        ha='center',
        va='bottom',
        fontsize=10
        )

plt.subplot(3, 1, 2)
bars = plt.bar(funnel_device['step'], funnel_device['users_web'])
plt.title('Step-by-step воронка: конверсия от предыдущего шага (%) для web версии')
plt.xlabel('Этапы воронки')
plt.ylabel('Конверсия от предыдущего шага в %')

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height}',
        ha='center',
        va='bottom',
        fontsize=10
        )

plt.show()
````

````python
br_df = df_hist[df_hist['country_code'] == 'BR']
mx_df = df_hist[df_hist['country_code'] == 'MX']
ar_df = df_hist[df_hist['country_code'] == 'AR']
co_df = df_hist[df_hist['country_code'] == 'CO']

dfs = {
    'Бразилия': br_df,
    'Мексика': mx_df,
    'Аргентина' : ar_df,
    'Колумбия' : co_df
}

for name, df in dfs.items(): 
    df_steps = df[df['event_name'].isin(funnel_steps)]
    counts = df_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)
    funnel = round((counts / counts.shift(1) * 100), 1)
    funnel_country = pd.DataFrame({
    'step': funnel_steps[1:],
    'users': funnel[1:]})
 
    plt.figure(figsize = (12, 6))

    #plt.subplot(3, 1, 1)
    bars = plt.bar(funnel_country['step'], funnel_country['users'])
    plt.title(f'Step-by-step воронка: конверсия от предыдущего шага (%) для {name}')
    plt.xlabel('Этапы воронки')
    plt.ylabel('Конверсия от предыдущего шага в %')

    for bar in bars:
        height = bar.get_height()
        plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height}',
        ha='center',
        va='bottom',
        fontsize=10
        )

    plt.show()
````

**Задача 2.3.** Влияние уровня риска актива на открытие второго депозита.
- Изучите, как уровень риска первого актива влияет на конверсию во второй депозит.

Нулевая гипотеза:
Конверсии в группах control и test равны:
pA = pB

Альтернативная гипотеза:
Конверсии в группе tets выше чем в группе control:
pA < pB

````python
alpha = 0.05 

results = []

for level in df_levels_dif_conversion['risk_level'].unique():
    data = df_levels_dif_conversion[df_levels_dif_conversion['risk_level'] == level]
    n_control = data['users_asset_purchase_control'].values[0]
    n_test = data['users_asset_purchase_test'].values[0]

    m_control = data['users_second_deposits_control'].values[0]
    m_test = data['users_second_deposits_test'].values[0]
    
    stat_ztest, p_value_ztest = proportions_ztest(
    [m_control, m_test],
    [n_control, n_test],
    alternative='smaller')
    
    results.append({
        'risk_level' : level,
        'p-value' : round(p_value_ztest, 4)})
    
    print(f'Значение p-value: {round(p_value_ztest, 4)}')
    
    if p_value_ztest > alpha:
        print(f'pvalue={p_value_ztest} > {alpha}')
        print(f'Нулевая гипотеза находит подтверждение для {level}!')
    else:
        print(f'pvalue={p_value_ztest} < {alpha}')
        print(f'Нулевая гипотеза не находит подтверждения! для {level}')
        
results_df = pd.DataFrame(results)
````

### Ключевая метрика - средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию)
#### Гипотезы:

- <b> H0 </b> : Средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию) одинакова и в контрольной, и в тестовой группе.

- <b> H1 </b> : Средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию) тестовой группе отлична от контрольной.

````python
_, p_value = ttest_ind(total_amount['test'], total_amount['control'], equal_var=False)
print(f'Значение p-value: {round(p_value, 4)}')

if p_value > alpha:
    print(f'p-value = {round(p_value, 4)} > {alpha} → различия не значимы')
else:
    print(f'p-value = {round(p_value, 4)} < {alpha} → различия статистически значимы')
````

### Барьерная метрика — конверсия из регистрации в первый депозит.

- <b> H0 </b>: Конверсия из регистрации в первый депозит одинакова и в контрольной, и в тестовой группе.

- <b> H1 </b>: Конверсия из регистрации в первый депозит в тестовой группе отлична от конверсии в контрольной.

````python
first_deposits = [first_deposit['test'], first_deposit['control']]
registration = [registrated['test'], registrated['control']]

z_stat, p_value = proportions_ztest(first_deposits, registration, alternative = 'two-sided')
if p_value > alpha:
    print(f'p-value = {round(p_value, 4)} > {alpha} → различия не значимы')
else:
    print(f'p-value = {round(p_value, 4)} < {alpha} → различия статистически значимы')
````

### Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- <b> H0 </b>: Конверсия из первого депозита во второй одинакова и в контрольной, и в тестовой группе.

- <b> H1 </b>: Конверсия из первого депозита во второй в тестовой группе отлична от конверсии в контрольной.

````python
first_deposits = [first_deposit['test'], first_deposit['control']]
second_deposits = [second_deposit['test'], second_deposit['control']]

z_stat, p_value = proportions_ztest(second_deposit, first_deposits, alternative = 'two-sided')
if p_value > alpha:
    print(f'p-value = {round(p_value, 4)} > {alpha} → различия не значимы')
else:
    print(f'p-value = {round(p_value, 4)} < {alpha} → различия статистически значимы')
````

### Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.
- <b> H0 </b>: Cредняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит одинакова и в контрольной, и в тестовой группе.

- <b> H1 </b>: Cредняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит в тестовой группе отлична от конверсии в контрольной.

````python
avg_arpu_first = df_abt[df_abt['event_name'] == 'first_deposit'].groupby(['user_id', 'group'])['amount'].sum().reset_index().fillna(0)
control = avg_arpu_first[avg_arpu_first['group'] == 'control']['amount']
test = avg_arpu_first[avg_arpu_first['group'] == 'test']['amount']

_, p_value = ttest_ind(test, control, equal_var=False)
print(f'Значение p-value: {round(p_value, 4)}')

if p_value > alpha:
    print(f'p-value = {round(p_value, 4)} > {alpha} → различия не значимы')
else:
    print(f'p-value = {round(p_value, 4)} < {alpha} → различия статистически значимы')
````
